# 主成分分析

## 特異値分解
実行列$\mathsf{H}$は，$\mathsf{H}=\mathsf{U}\mathsf{S}\mathsf{V}^T$の形に分解できる．
但し，$\mathsf{U}$, $\mathsf{V}$は正規直交行列，$\mathsf{S}$は対角行列．

特異値分解を用いると
$\displaystyle \boldsymbol{x}_i=\sum_{j=1}^m=\lambda_jv_{ji}\boldsymbol{u}_j$
のように，各データ$\boldsymbol{x}_i$が共通の基底$\{\boldsymbol{u}_j\}$による展開が得られる．

```python
import numpy as np
import csv
import matplotlib.pylab as plt

# データの読み込み
csvfile = open( "rain2016.csv", 'r' )
csvin = csv.reader(csvfile)

n = 0
for row in csvin:
  if n == 0:
    header = row
    m = len(row)
  elif n == 1:
    xdata = np.reshape(np.array(row).astype("float"), (1,m))
  else:
    xtemp = np.reshape(np.array(row).astype("float"), (1,m))
    xdata = np.append( xdata, xtemp, axis=0 )
  n += 1

n = n - 1

# 特異値分解
xd = (xdata - np.mean(xdata,0)).T
U,a,Vt = np.linalg.svd( xd )
```

```python
plt.plot(U[:,0])
plt.plot(U[:,1])
plt.plot(U[:,2])
plt.show()
```

```python
xmon = np.arange(1,13)

plt.plot(xmon,Vt[0,:])
plt.plot(xmon,Vt[1,:])
plt.plot(xmon,Vt[2,:])
plt.xticks(range(0,15,3))
plt.show()
```
